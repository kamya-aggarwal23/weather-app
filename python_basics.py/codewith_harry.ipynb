# CHAPTER 3 – STRINGS

A string is a data type in Python that represents a sequence of characters enclosed in quotes.

Strings can be written using:

* Single quotes (' ')
* Double quotes (" ")
* Triple quotes (''' ''')

## STRING SLICING

String slicing is used to access a part of a string.

* Indexing starts from 0 and goes up to (length - 1).
* Negative indexing starts from -1 for the last character.
* In slicing, the starting index is included and the ending index is excluded.

### Negative Indices

* -1 refers to the last character.
* -2 refers to the second last character.
* Negative indexing continues similarly for the remaining characters.

For beginner to intermediate Python, these are the **most commonly used string functions**:

### String Functions

* `lower()` – Converts to lowercase
* `upper()` – Converts to uppercase
* `capitalize()` – Capitalizes first letter
* `title()` – Capitalizes first letter of each word
* `strip()` – Removes spaces from both ends
* `replace()` – Replaces a substring
* `find()` – Finds index of first occurrence
* `count()` – Counts occurrences of a substring
* `startswith()` – Checks if string starts with given text
* `endswith()` – Checks if string ends with given text

# Chapter 4: Lists and Tuples

## Lists in Python
* A list is a collection of items.
* Lists can store values of different data types.
* Lists are **mutable** (can be modified).
* Created using **[ ]**.

### List Operations
* **Concatenation (+)** → Joins two lists.
* **Repetition (*)** → Repeats list elements.
* **Membership (in, not in)** → Checks if an element exists.
* **Length (len())** → Returns number of elements.
* **Slicing** → Accesses a range of elements.

### Important List Methods
* **append(x)** → Adds element at the end.
* **extend(iterable)** → Adds multiple elements from another list.
* **insert(i, x)** → Inserts element at index `i`.
* **remove(x)** → Removes first occurrence of `x`.
* **pop(i)** → Removes and returns element at index `i`.
* **clear()** → Removes all elements.
* **sort()** → Sorts list in ascending order.
* **reverse()** → Reverses list order.
* **count(x)** → Counts occurrences of `x`.
* **index(x)** → Returns index of first occurrence of `x`.
* **copy()** → Creates a copy of the list.
---

# Tuples in Python
* A tuple is a collection of items.
* Tuples can store different data types.
* Tuples are **immutable** (cannot be modified).
* Created using **( )**.

### Tuple Operations
* **Concatenation (+)** → Combines tuples.
* **Repetition (*)** → Repeats tuple elements.
* **Membership (in, not in)** → Checks if an element exists.
* **Length (len())** → Returns number of elements.
* **Slicing** → Accesses a range of elements.

### Tuple Methods
> Tuples have only **two built-in methods**
* **count(x)** → Returns number of times `x` occurs.
* **index(x)** → Returns index of first occurrence of `x`.

### Useful Functions with Tuples
These are functions, not tuple methods:
* **len(tuple)** → Number of elements.
* **max(tuple)** → Largest element.
* **min(tuple)** → Smallest element.
* **sum(tuple)** → Sum of elements.
* **sorted(tuple)** → Returns a sorted list.
* **tuple(list)** → Converts a list into a tuple.

# Chapter 7: Loops in Python

## Loops
* Loops are used to execute a block of code repeatedly.
* They reduce repetition and make programs shorter and easier to write.

### Types of Loops
1. **while loop**
2. **for loop**
---

## While Loop
* Repeats a block of code as long as a condition is True.
* Condition is checked before each iteration.
* Stops when the condition becomes False.
* If the condition never becomes False, it creates an **infinite loop**.
---

## For Loop
* Used to iterate through a sequence.
* Can be used with strings, lists, tuples, sets, and dictionaries.
* Executes once for each item in the sequence.
---

## Range Function
* Used to generate a sequence of numbers.
* Commonly used with for loops.
* Can specify:

  * Start value
  * Stop value
  * Step size
---

## For Loop with Else
* An optional **else** block can be used with a loop.
* The else block executes when the loop finishes normally.
* It does not execute if the loop is terminated using **break**.
---

## Break Statement
* Used to immediately exit a loop.
* Control moves to the statement after the loop.
---

## Continue Statement
* Skips the current iteration.
* Continues execution with the next iteration of the loop.
---

## Pass Statement
* A null statement in Python.
* Used when a statement is required syntactically but no action is needed.
* Means "do nothing".

# Chapter 8: Functions and Recursion

## Functions
* A function is a group of statements that performs a specific task.
* Functions help avoid repetition of code.
* A function can be reused multiple times in a program.

### Function Definition
* The set of instructions written inside a function.
* Defines what the function will do.

### Function Call
* Used to execute a function.
* A function runs only when it is called.

---

## Types of Functions

### Built-in Functions
* Functions already provided by Python.
* Examples: `print()`, `len()`, `range()`, `type()`.

### User-defined Functions
* Functions created by the programmer.
* Used to perform specific tasks as needed.

---

## Functions with Arguments
* Arguments are values passed to a function.
* They allow the function to work with different data.
* A function can accept zero, one, or multiple arguments.

### Return Value
* A function can return a result back to the caller.
* Returned values can be stored and used later.

---

## Default Parameter Value
* A parameter can have a default value.
* The default value is used when no argument is provided.
* If an argument is passed, it overrides the default value.

---

## Recursion
* Recursion is a process where a function calls itself.
* Useful for solving problems that can be broken into smaller similar problems.
* Commonly used in factorial, Fibonacci series, tree traversal, etc.

### Base Case
* A stopping condition in recursion.
* Prevents infinite recursive calls.
* Every recursive function must have a base case.

# Chapter 9 – File I/O (Notes)

## 1. What is File I/O?
* RAM is **volatile memory**, so data is lost when a program stops.
* To store data permanently, we use **files**.
* A file is data stored on a storage device (HDD/SSD).
* Python programs can:

  * **Read** data from files.
  * **Write** data to files.

### Memory Types
* **RAM** → Volatile (temporary)
* **HDD/SSD** → Non-volatile (permanent)
---
## 2. Types of Files

### Text Files
* Store data as readable text.
* Examples:

  * `.txt`
  * `.c`
  * `.py`

### Binary Files
* Store data in binary format.
* Examples:

  * `.jpg`
  * `.dat`
  * `.mp3`
---
## 3. Opening a File
* Files must be opened before reading or writing.
* Python provides the **open()** function.
* It requires:

  1. Filename
  2. Mode

---
## 4. Reading a File
* Reading means retrieving data from a file.
* The entire content of a file can be read at once.
* After reading, the file should be closed.

### Reading One Line
* A file can also be read **one line at a time**.

---

## 5. Modes of Opening a File

| Mode | Purpose                                     |
| ---- | ------------------------------------------- |
| `r`  | Open for reading                            |
| `w`  | Open for writing                            |
| `a`  | Open for appending (adding data at the end) |
| `+`  | Open for updating (read and write)          |
| `rb` | Read in binary mode                         |
| `rt` | Read in text mode                           |
---
## 6. Writing to a File
* To write data, open the file in:

  * **Write mode (`w`)**, or
  * **Append mode (`a`)**
* Writing allows a program to store information in a file permanently.
### Write Mode (`w`)
* Creates a new file if it does not exist.
* Overwrites existing content if the file already exists.
### Append Mode (`a`)
* Adds new content at the end of the file.
* Existing content remains unchanged.
---
## 7. Closing a File

* Files should be closed after use.
* Closing a file frees system resources and saves changes properly.
---
## 8. With Statement
* The **with statement** is the preferred way to work with files.
* It automatically closes the file after use.
* Safer and cleaner than manually closing files.
### Advantages
* Automatic file closing.
* Cleaner code.
* Reduces chances of errors.

# Chapter 10 – Object Oriented Programming (OOP) Notes

## 1. Object-Oriented Programming (OOP)
* OOP is a programming approach where problems are solved by creating **objects**.
* It focuses on **reusable code** (DRY Principle – Don't Repeat Yourself).
* Real-world entities are represented using classes and objects.
---

# 2. Class
* A **class** is a blueprint or template for creating objects.
* It contains information needed to create valid objects.
* A class defines:
  * Attributes (data)
  * Methods (functions)

### Example Analogy
* Blank Form → Class
* Filled Form → Object

A class itself does not occupy memory for individual data. Memory is allocated when an object is created.
---

# 3. Object
* An **object** is an instance of a class.
* Objects are created from classes.
* Each object has its own data and can use the methods defined in the class.

### Important Points
* Class = Blueprint
* Object = Actual thing created from blueprint
---

# 4. Modelling a Problem in OOP

When solving a problem using OOP:

### Nouns → Classes

Example:
* Employee
* Student
* Car

### Adjectives → Attributes

Example:
* Name
* Age
* Salary
* Color

### Verbs → Methods

Example:
* Get Salary
* Increment Salary
* Display Details
---

# 5. Class Attributes
* Attributes that belong to the **class itself**.
* Shared by all objects of that class.
* Only one copy exists regardless of the number of objects.

### Characteristics
* Common to every object.
* Can be accessed using the class name.
* Can be modified at the class level.
---

# 6. Instance Attributes
* Attributes that belong to a specific object (instance).
* Different objects can have different values.

### Characteristics
* Stored separately for each object.
* Created specifically for an object.

### Priority Rule
When Python looks for an attribute:
1. Check the object (instance) first.
2. If not found, check the class.

Therefore:
**Instance Attributes take precedence over Class Attributes.**
---

# 7. Self Parameter
* `self` refers to the current object (instance) of a class.
* It is automatically passed when a method is called through an object.
* It allows methods to access instance attributes and other methods.

### Important Points
* `self` represents the current object.
* It must be the first parameter in instance methods.
* Python automatically supplies it.
---

# 8. Static Method

Sometimes a method does not need access to:
* Instance attributes
* Class attributes

Such methods are called **Static Methods**.
### Characteristics
* Do not use `self`.
* Belong to the class.
* Used for utility functions.

### Use Cases
* Greeting messages
* Calculations
* Helper functions
---

# 9. Constructor (**init**)

### Definition
* `__init__()` is a special method.
* It runs automatically when an object is created.
* Also called a **Constructor**.

### Purpose
* Initialize object data.
* Assign values to attributes during object creation.

### Characteristics
* Executed automatically.
* Can accept arguments.
* Uses `self`.

````python
class Employee:
    language = "Python"      # Class attribute (shared by all objects)
    salary = 1200000         # Class attribute

    # Constructor (__init__) runs automatically when an object is created
    def __init__(self, name, salary, language):
        self.name = name          # Instance attribute
        self.salary = salary      # Instance attribute
        self.language = language  # Instance attribute

        print("I am creating an object")

    # Instance method
    def getInfo(self):
        print(f"The language is {self.language}. The salary is {self.salary}")

    # Static method (does not use self)
    @staticmethod
    def greet():
        print("Good morning")


# Creating an object named harry
harry = Employee("Harry", 1300000, "JavaScript")

# Printing instance attributes
print(harry.name, harry.salary)

# Creating another object
rohan = Employee("Rohan", 1000000, "Python")
````

# Chapter 11 – Inheritance & More on OOP (Brief Notes)

## 1. Inheritance

* Inheritance is the process of creating a new class from an existing class.
* The new class (child/derived class) gets access to the attributes and methods of the existing class (parent/base class).
* Promotes **code reusability** and reduces duplication.

### Benefits

* Reuse existing code.
* Easier maintenance.
* Better organization of related classes.

### Important

* A child class can:

  * Use parent methods and attributes.
  * Add new methods and attributes.
  * Override existing methods and attributes.

---

# 2. Types of Inheritance

### Single Inheritance

* One child class inherits from one parent class.

**Parent → Child**

---

### Multiple Inheritance

* One child class inherits from multiple parent classes.

**Parent1 + Parent2 → Child**

### Important

* Python supports multiple inheritance directly.
* If two parents have the same method, Python follows the **Method Resolution Order (MRO)**.

---

### Multilevel Inheritance

* A child class becomes a parent of another class.

**Parent → Child1 → Child2**

Child2 inherits features from both Child1 and Parent.

---

# 3. super() Method

* Used to access methods and constructors of the parent class.
* Commonly used inside constructors.

### Why use it?

* Avoid rewriting parent initialization code.
* Access parent methods when they are overridden.

### Remember

* `super()` refers to the immediate parent class.

---

# 4. Class Methods

* A class method works with the **class itself**, not individual objects.
* Created using the `@classmethod` decorator.

### Uses

* Modify class attributes.
* Create alternative constructors.
* Perform operations related to the whole class.

### Difference

| Method Type     | Works With                    |
| --------------- | ----------------------------- |
| Instance Method | Object                        |
| Class Method    | Class                         |
| Static Method   | Neither object nor class data |

---

# 5. Property Decorators

Property decorators allow methods to behave like attributes.

### Why use them?

* Hide implementation details.
* Make code cleaner.
* Control access to attributes.

Instead of calling a method like a function, it can be accessed like a normal attribute.

---

# 6. Getters

* A getter retrieves the value of an attribute.
* Used when you want to calculate or control what is returned.

### Purpose

* Encapsulation
* Data protection
* Computed attributes

---

# 7. Setters

* A setter updates the value of an attribute.
* Allows validation before assigning data.

### Benefits

* Prevent invalid values.
* Keep object data consistent.
* Add custom logic during assignment.

---

# 8. Operator Overloading

Python allows operators to work with user-defined objects.

### Example Concept

Normally:

* `+` adds numbers.
* `-` subtracts numbers.

With operator overloading, these operators can work on your own classes.

### Common Overloadable Operators

| Operator | Meaning        |
| -------- | -------------- |
| +        | Addition       |
| -        | Subtraction    |
| *        | Multiplication |
| /        | Division       |
| //       | Floor Division |

---

# 9. Dunder (Magic) Methods

* Special methods surrounded by double underscores.
* Automatically called by Python.

Examples:

* Constructor
* String representation
* Operator overloading methods

### Why Important?

* They make objects behave like built-in types.

---

# Project 2 – The Perfect Guess

We are going to write a program that generates a random number and asks the user to guess it.

* If the player's guess is **higher** than the actual number, the program displays:

  **"Lower number please"**

* If the player's guess is **lower** than the actual number, the program displays:

  **"Higher number please"**

* When the user guesses the correct number, the program displays the **number of guesses** the player used to arrive at the correct answer.

### Hint

Use the **random** module.


````python
from random import randint

num = randint(0, 100)
guess = -1
i = 0

while num != guess:
    guess = int(input("Enter a number: "))
    i += 1

    if guess > num:
        print("Lower number please!")
    elif guess < num:
        print("Higher number please!")
print("CONGRATULATIONS!!")
print(f"You guessed the correct answer {num} in {i} guesses.")
````
